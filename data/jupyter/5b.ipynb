### Executive Summary of Findings

The regression analysis provides strong, significant support that firms with more tangible assets (collateral), higher effective tax rates (debt shield), and larger size (lower distress risk) all have significantly higher leverage.

The model also produced one anomalous finding: Log_Market_to_Book is positively and significantly related to leverage, which contradicts the initial expected effect.

### Detailed Interpretation of Coefficients

#### Here is a variable-by-variable breakdown in economic terms.

#### Group 1: Statistically Significant Coefficients

These are the variables where the model found a reliable, non-zero effect.

1. Tangibility (Coefficient: +0.3458)

Direction: Positive.

Magnitude & Units: This is the strongest and most economically significant finding. A 1-unit increase in the tangibility ratio (i.e., going from 0% to 100% of assets being tangible) is associated with a 0.3458-unit (or 34.58 percentage point) increase in the leverage ratio, holding all other factors constant.

Simpler Terms: A 10 percentage point increase in a firm's asset tangibility (e.g., from 20% to 30%) is associated with a 3.46 percentage point increase in its leverage.

Economic Meaning: Tangible assets serve as good collateral, which reduces risk for lenders and lowers the firm's cost of debt. Firms with more collateral, therefore, take on significantly more debt.

2. Effective Tax Rate (Coefficient: +0.08)

Direction: Positive.

Magnitude & Units: A 1-unit increase in the effective tax rate (from 0% to 100%) is associated with a 0.08-unit (or 8 percentage point) increase in the leverage ratio, ceteris paribus.

Simpler Terms: A 10 percentage point increase in a firm's tax rate (e.g., from 21% to 31%) is associated with a 0.8 percentage point increase in its leverage.

Economic Meaning: The debt tax shield is more valuable to firms in higher tax brackets, giving them a greater incentive to use debt financing. The magnitude, however, appears modest compared to tangibility.

3. Log_Total_Assets (Coefficient: +0.0139)

Direction: Positive.

Magnitude & Units: This is a log-lin model, so the interpretation is different. A 1% increase in a firm's Total_Assets is associated with a (0.0139 / 100)-unit increase in leverage, or a 0.0139 percentage point increase.

Simpler Terms: To see a meaningful change, a 100% increase in firm size (i.e., the firm's assets double) is associated with a 1.39 percentage point increase in its leverage ratio.

Economic Meaning: Larger firms are typically more diversified, have more stable cash flows, and have a lower risk of bankruptcy. This lower "cost of financial distress" allows them to safely take on more debt.

4. Log_Market_to_Book (Coefficient: +0.026) - [Anomalous Finding]

Direction: Positive. This result is statistically significant but contradicts the initial expected effect, which predicted a negative sign.

Magnitude & Units: As a log-lin variable, a 1% increase in the Market_to_Book ratio is associated with a 0.026 percentage point increase in the leverage ratio.

Simpler Terms: A 100% increase in a firm's M/B ratio (e.g., from 2.0 to 4.0) is associated with a 2.6 percentage point increase in its leverage ratio.

Economic Meaning: Theory suggests high-growth firms (high M/B) would use less debt to avoid risking valuable, intangible growth opportunities. The model's positive result suggests an alternative story: perhaps firms with high M/B ratios are viewed favorably by the market and have better access to all forms of capital, including debt, allowing them to borrow more easily.

6. Realized_Volatility (Coefficient: Insignificant)

Interpretation: The model found no statistically significant relationship between a firm's stock volatility (a proxy for risk) and its leverage.

Economic Meaning: This result fails to support that firms with higher business risk (and thus higher potential distress costs) will use less debt.

7. Turnover (Coefficient: Insignificant)

Interpretation: The model found no statistically significant relationship between a firm's stock liquidity (Turnover) and its leverage.

Economic Meaning: This fails to support the hypothesis that firms with more liquid stock (which makes it cheaper to issue equity) would substitute away from debt financing.
