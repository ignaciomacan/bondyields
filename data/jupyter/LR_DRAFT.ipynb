## Introduction

&nbsp;&nbsp;&nbsp;&nbsp; In finance, few metrics reveal as much about a company’s financial health and behavior as its leverage ratio, the extent to which a firm relies on debt to finance its assets $$Leverage Ratio = (Total Debt/Total Assets)$$
&nbsp;&nbsp;&nbsp;&nbsp;Leverage is an instrument of tradeoffs as it amplifies both gains and losses. A highly leveraged firm can achieve higher returns when times are good, but it also faces magnified risk when revenues fall. Because of this, leverage ratios are closely watched indicators of financial stability, investor confidence, and broader economic health. Recent evidence shows that corporate credit cycles in particular, are deeply intertwined with macroeconomic stability, even more than household credit. Expansions in corporate debt are associated with higher probabilities of future financial crises, slower recoveries, and greater financial sector distress (Ivashina, Kalemli-Özcan, Laeven, & Müller, 2024). Periods of widespread over leveraging have historically preceded financial crises, while shifts toward deleveraging can signal growing caution across industries. Analysts, policymakers, and investors alike track leverage trends not just to assess individual firm risk, but to gauge systemic vulnerability in the economy. In this way, leverage serves as a lens through which we can understand both corporate capital structure and use it's macroeconomic predictor power.

Building on that intuition, this project asks a focused question: **What firm-specific characteristics explain the variation in corporate leverage among U.S. firms?** In other words, what drives differences in how companies balance debt and equity, and what signals might foreshadow those shifts? Prior research has examined similar relationships over time, particularly the role of asset structure in shaping leverage decisions. For example, Norden and van Kampen (2013) analyzed U.S. firms from 1990 to 2010 and found that tangible assets such as property, plant, and equipment are major determinants of leverage through what they describe as the collateral channel. Their time series evidence highlights how the composition of corporate assets influences borrowing capacity and financial constraints, especially during periods of financial stress. While our study shares this motivation, it focuses on a cross-sectional snapshot, a single period of firm balance sheet and income statment data, to identify which financial characteristics are most predictive of leverage levels at a given point in time. By doing so, we aim to provide an updated view of the determinants of corporate leverage, capturing how firms differ in their financing choices within the current economic environment.



## Literature Review

&nbsp;&nbsp;&nbsp;&nbsp; The principle of the research question we are proposing calls on a widely discussed and studied claim that the corporate capital structure of a firm (to what degree the firm depends on debt over assets) can be predicted given certain characterisitcs a firm's balance sheet. The study of corporate capital structure is the attempt to understand why firms choose different mixtures of debt and equity to finance their operations. In a practical sense, a firm’s reliance on debt serves as a measure of the risk it assumes, sometimes a glaring red flag for a company that might otherwise seem healthy. One foundational work in this area is Modigalian and Miller's "Cost of Capital, Corporation Finance and the Theory of Investment" (Modigliani & Miller, 1958). The paper provided the M&M theorem, which proposed that in a "perfect" market (meaning there is no taxes, no bankruptcy costs, and no asymmetric information), a firm's value is evidently independent of its capital structure. The "capital structure irrelevance proposition" that the theorem suggested created confusion, as firms in the real world clearly do spend significant effort managing their debt-to-asset ratios. However, the M&M theorem's true value was in identifying the "frictions" that do make capital structure relevant. Things such as: taxes, financial distress costs, and information asymmetries.
 
&nbsp;&nbsp;&nbsp;&nbsp; From these implied "frictions" identified by the M&M theorem, two primary theories emerged. The first being the Trade-Off Theory. Again from the same economists, Modigliani & Miller (1963), relaxing the M&M tax assumption shows that due to the interest payments on debt being tax-deductible, debt creates a "tax shield" that adds value to the firm. This would suggest firms should be 100% debt-financed. However, this is balanced by the costs of financial distress as leverage (debt) increases, as does the probability of bankruptcy, which incurs substantial legal and reputational costs. The Trade-Off Theory provides that an optimal capital structure exists where the marginal benefit of the debt tax shield is exactly offset by the marginal cost of potential financial distress.

&nbsp;&nbsp;&nbsp;&nbsp; The second dominant theory is the Pecking Order Theory, which was formalized by Myers and Majluf (1984) in "Corporate financing and investment decisions when firms have information that investors do not have." This Pecking Order theory is not based on a static trade-off, rather on information asymmetry, which is the idea that managers know more about the firm's true value and prospects than outside investors. With this considered, raising capital sends signals to the market. If a firm issues equity, investors may assume the stock is overvalued, which would subsquently cause the equity share price to fall. To avoid this, managers will follow a "pecking order" of financing preferences: they first use internal funds or retained earnings, which sends no signal. In the case that external funds are required, they issue the "safest" security first, which is debt, while equity is issued only as a last resort.

| **Turnover (Liquidity)** | **−** | High stock liquidity lowers the cost of issuing equity. Since firms trade off the costs of debt vs. equity, a lower equity cost will lead them to use more equity and, consequently, less debt. |

<font color="maroon" size="6"><b>1.</b></font> **Leverage Ratio (Dependent Variable)**  

From the balance sheet, we extracted two key variables:  
- **Total Debt** 
- **Total Assets**  

These are used to calculate the firm’s leverage ratio as follows:

$$
\text{Leverage Ratio} = \frac{\text{Total Debt}}{\text{Total Assets}}
$$

<font color="maroon" size="6"><b>2.</b></font> **Return on Assets (ROA)**  


The first explanatory variable is, **Return on Assets (ROA)**, which measures a firm's profitability relative to its total assets.  
This variable captures how efficiently a firm uses its asset base to generate net income.

- **Net Income** 
- **Total Assets**

Both statements were transposed to align reporting periods, and we selected the same quarter used for the dependent variable.

The ROA is then calculated as:

$$
\text{ROA} = \frac{\text{Net Income}}{\text{Total Assets}}
$$


<font color="maroon" size="6"><b>3.</b></font> **Asset Tangibility (PPE / Total Assets)**  


The next explanatory variable measures the **tangibility of assets**, which captures the proportion of a firm's total assets that are physical and can be used as collateral.  
In capital structure theory, firms with more tangible assets generally have greater access to debt financing, since these assets can serve as security for lenders.

- **Net Property, Plant, and Equipment (Net PPE)**
- **Total Assets**

As with the previous variable, statements were transposed to align reporting periods.

The tangibility ratio is then calculated as:

$$
\text{Tangibility} = \frac{\text{Net PPE}}{\text{Total Assets}}
$$

<font color="maroon" size="6"><b>4.</b></font> **Firm Size (Log of Total Assets)**  


Our next explanatory variable captures **firm size**, a key determinant of financial structure.  
Larger firms generally have more diversified operations, lower default risk, and better access to credit markets, which can influence their leverage decisions.

Because total assets vary by several orders of magnitude across firms, taking the logarithm reduces scale differences, mitigates heteroskedasticity, and linearizes the relationship between firm size and leverage.

$$
\text{Firm Size} = \ln(\text{Total Assets})
$$


<font color="maroon" size="6"><b>5.</b></font> **Market-to-Book Ratio (M/B)**  


Next the **Market-to-Book Ratio (M/B)**, captures how investors value a firm relative to its accounting (book) value.  
A higher M/B ratio indicates that the market expects greater growth potential or intangible value not reflected on the balance sheet, while a lower ratio suggests the market values the firm closer to its tangible assets.

- **Book Equity (Common Stock Equity)**
- **Shares Outstanding**
- **Stock Price**

The **Market Value of Equity (Market Capitalization)** is calculated as the product of the share price and shares outstanding.  
We then divide this market value by book equity to obtain the Market-to-Book ratio.  
Finally, a log is applied to reduce skewness and make the variable more normally distributed for regression analysis.

The Market-to-Book formula is:

$$
\text{Market-to-Book} = \frac{\text{Price} \times \text{Shares Outstanding}}{\text{Book Equity}}
$$
Then
$$
\ln(\text{Market/Book})
$$

<font color="maroon" size="6"><b>6.</b></font> **Effective Tax Rate (ETR)**  


The **Effective Tax Rate (ETR)** measures the percentage of a firm’s pre-tax income that is paid in taxes.  
It provides insight into how taxation affects a company’s net profitability and can vary due to tax credits, foreign income, or deferred tax adjustments.  

This variable reflects the company’s average tax burden as reported in its quarterly financials.  

The formula for the effective tax rate is:

$$
\text{ETR} = \frac{\text{Tax Expense}}{\text{Income Before Tax}}
$$

<font color="maroon" size="6"><b>7.</b></font> **Stock Price Volatility (Realized Volatility)**  

The **Realized Volatility** variable captures fluctuations in a firm’s stock price over the quarter and serves as a proxy for **market uncertainty** or **risk perception**.  
In capital structure theory, firms with higher volatility face greater financial uncertainty and are less likely to use high levels of debt, while firms with stable prices may take on more leverage due to their lower perceived risk of default.

We first compute the **daily returns** by looking at closing prices between days of the period, then take their **standard deviation** to obtain the realized volatility for the quarter.  

This value reflects the average daily percentage change in the firm’s stock price.  

The Realized Volatility formulas are:

$$
r_t = \frac{P_t - P_{t-1}}{P_{t-1}}
$$

$$
\sigma_{quarterly} = \sqrt{\frac{1}{n-1}\sum_{t=1}^{n}(r_t - \bar{r})^2}
$$

<font color="maroon" size="6"><b>8.</b></font> **Stock Turnover (Trading Activity)**  


The **Turnover Ratio** measures how frequently a firm’s shares change hands in the market and serves as a proxy for **liquidity** and **investor activity**.  
A higher turnover ratio indicates that a larger proportion of the company’s outstanding shares are being traded on an average day, reflecting **greater liquidity**, **market attention**, and **ease of trading**.  
Conversely, a lower turnover ratio may signal **less liquidity**, **higher insider ownership**, or a **smaller free float** (fewer shares available to trade).  

Turnover connects to capital structure decisions through **market liquidity and information asymmetry**.  
Firms with highly liquid stocks (high turnover) generally face lower transaction costs and easier access to external financing, while firms with illiquid stocks (low turnover) may face higher costs of capital and rely more on internal funding.  

To calculate this variable, we divide a firm’s **average daily trading volume** over the quarter by its **total shares outstanding** as reported on the balance sheet.  
This ratio expresses what fraction of the firm’s equity base is traded each day on average.

The Turnover Ratio formula is:

$$
\text{Turnover} = \frac{\text{Average Daily Trading Volume}}{\text{Shares Outstanding}}
$$

# Limitations, Policy Implications, and Extensions

## Discuss key limitations of your analysis (data, methodology, scope)

Methodological Limitation (Cross-Sectional OLS): Considering nature of the model being a cross-sectional OLS regression, the primary limitation here is that capital structure decisions are dynamic, rather than static. A firm's leverage today is heavily influenced by its leverage from last year and its long-term "target" leverage. A static OLS model cannot capture this adjustment process.

Data Limitation (Samples): The samples consists of firms only listed on the NASDQ. These are, by definition, large, mature, successful, and publicly traded. This sample selection bias means your findings are not representative of all firms. Small firms, private firms, and startups face entirely different (and often more severe) financing constraints.

Scope Limitation (Firm-Specific Factors): The model only includes firm-specific, or micro characteristics. It completely omits macroeconomic factors that can heavily influence financing decisions, such as prevailing interest rates, credit spreads, GDP growth, and inflation.

## Identify potential sources of bias and how they might affect conclusions

The limitations above introduce several potential sources of bias, which could affect the validity of your coefficients:

Endogeneity: This is the most significant econometric challenge in corporate finance. The OLS assumes the explanatory variables (e.g., ROA) cause the dependent variable (Leverage), but the relationship could potentially be a two-way relationship.
            OLS cannot distinguish between the two effects if the do in fact cause each other, leading to a biased and inconsistent coefficient for β (ROA).

Omitted Variable Bias (OVB): The model omits key variables that are correlated with both the independent variables and your dependent variable.

Measurement Error: The model variables are proxies for theoretical concepts, and they may be imperfect.

## Describe policy implications of your findings

For Corporate Managers (CFOs): 

If β(TaxRate) is positive and significant, it provides evidence that the debt tax shield is a primary driver of financing policy. This validates the managerial focus on using debt to reduce corporate tax burdens. Conversely, if β(Volatility) is negative and significant, it serves as a strong warning that managers of firms with high business risk should maintain financial flexibility by using less debt, as the market (and theory) links high risk to higher distress costs.

For Government & Regulators: 

The significance of the β(TaxRate) coefficient is a direct message to policymakers. It implies that changes in corporate tax policy have a direct, measurable effect on the financial risk of the corporate sector. For example, a policy that lowers the corporate tax rate would, according to this model, reduce the incentive for firms to use debt, potentially leading to a more stable, less-leveraged corporate economy.

Based from the key limitations discussed previously, there are two major extentions for future research on this research topic.

1. Adopt a Panel Data Methodology: The most important extension is to move from a cross-section to a panel dataset (many firms over many years). This would allow the data to:
Include firm fixed effects to control for all unobserved, time-invariant factors, which would significantly reduce omiited variabnle bias.

2. Incorporate Macroeconomic Variables: Extend the model to include significant macroeconomic time-series variables like the 10-year Treasury rate (as a proxy for the risk-free rate) or the Baa-Aaa credit spread (as a proxy for the cost of debt). This would test whether macro conditions are more important than firm-specific factors.

3. Conduct a Comparative Study: Re-run the analysis on different samples that do not share an underlying commonality. This model only considered companies publicly traded that are in the NASDAQ index, whereas data covering private firms and other public firms outside of the NASDAQ would have brought a greater data population.
